```python
def areaof_circle(r):
    return m.pi * r * r
def areaof_ractangle(l,b):
    return l*b
def areaof_square(s):
    return s*s
def areaof_triangle(b,h):
    return 0.5*b*h
def permiter_of_circle(r):
    return 2*m.pi*r


print(f"the area of the circle is: {areaof_circle(5)}")
print(f"the area of the rectangle is: {areaof_ractangle(5,10)}")
print(f"the area of the square is: {areaof_square(5)}")
print(f"the area of the triangle is: {areaof_triangle(5,10)}")
```

```python
# factorial of a number in for loop 
def factorial(n):
    for i in range(n-1,0,-1):
        n=n*i
    return n
print(f"the factorial of 5 is: {factorial(4)}")
```

```python
# factorial of a number in recursion in if condition
def fect(n):
    if n==0 or n==1:
        return 1
    else:
        n=n*fect(n-1)
        return n
n=int(input("enter the num: "))
print(f"the factorial of {n} is: {fect(n)}")
         
```

```python
def checkprime(n):
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True
n=[1,23,17,9,7,21,3,5,10,4,18]
l1=[]
l2=[]
for i in n: 
    if checkprime(i):
        l1.append(i)
    else:
        l2.append(i)
print(f"this a prime list {l1}")
print(f"this a non prime list {l2}")
```

```python
import matplotlib.pyplot as plt
over=range(1,11)
ind_scrore=[6,14,30,35,43,45,55,61,74,80]
nz_scrore=[5,15,19,24,36,42,55,51,59,65]
plt.plot(over, ind_scrore, "bo-", label="india")
plt.plot(over, nz_scrore, "ko-", label="new zealand")
plt.title("t20 world cup cricket score")
plt.xlabel("<-----over---->") # x-axis label
plt.ylabel("<-----------score--------->") # y-axis label
plt.xticks(over) # x-axis ticks user 
plt.legend() # legend use for show the label in graph
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as npi
npi.random.seed(1)
votes = npi.random.randint(100, 1000, 6)
name = ["giridharan", "pawan", "monish", "velan", "gokul J", "akash"]
plt.pie(votes, labels=name, autopct="%.2f%%", explode=[0, 0, 0, .2, .1, 0]) # autopct use for show the percentage in pie chart
plt.title("vote percentage")
plt.show()


```

```python
# bar graph 
import random

import matplotlib.pyplot as plt

colors = [f"#{random.randint(0, 0xFFFFFF):06x}" for _ in range(len(name))]
plt.bar(name, votes, color=colors)
plt.show
```
