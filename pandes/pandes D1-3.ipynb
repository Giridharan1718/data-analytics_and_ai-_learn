## Pandas: Detailed Explanation

**Pandas** is an essential Python library for data analysis and manipulation. Here are some key details:

### 1. Data Structures
- **Series**: A one-dimensional labeled array capable of holding any data type (integers, strings, floats, etc.).
- **DataFrame**: A two-dimensional, size-mutable, and heterogeneous tabular data structure with labeled axes (rows and columns).

### 2. Key Features
- **Data Cleaning**: Handle missing data, remove duplicates, and perform data type conversions.
- **Data Selection**: Easily select, filter, and slice data using labels or boolean indexing.
- **Aggregation and Grouping**: Summarize data using functions like `sum()`, `mean()`, `groupby()`, etc.
- **Merging and Joining**: Combine multiple datasets using functions like `merge()`, `join()`, and `concat()`.
- **Time Series Support**: Powerful tools for working with time-indexed data.
- **Input/Output**: Read and write data from various file formats (CSV, Excel, SQL, JSON, etc.).

### 3. Usage in Data Science
- **Exploratory Data Analysis (EDA)**: Quickly explore and visualize datasets.
- **Data Preprocessing**: Prepare data for machine learning models.
- **Statistical Analysis**: Perform descriptive and inferential statistics.

### 4. Example Workflow
1. **Import pandas**: `import pandas as pd`
2. **Load data**: `df = pd.read_csv('data.csv')`
3. **Inspect data**: `df.head()`, `df.info()`
4. **Clean data**: `df.dropna()`, `df.fillna()`
5. **Analyze data**: `df.groupby('column').mean()`
6. **Visualize data**: `df.plot()`

Pandas simplifies complex data tasks, making it a cornerstone tool for anyone working with data in Python.

### DataFrames in Pandas

A **DataFrame** is a 2D labeled data structure in Pandas, similar to a table with rows and columns.

- Columns can store different data types.
- You can create a DataFrame from dictionaries, lists, NumPy arrays, or files (CSV/Excel).
- Common operations include:
    - viewing data (`head()`, `tail()`, `info()`)
    - selecting columns/rows (`df['col']`, `loc`, `iloc`)
    - filtering (`df[df['col'] > value]`)
    - aggregating (`mean()`, `sum()`, `groupby()`)

DataFrames are the core structure used for analysis, cleaning, and transformation of tabular data in Pandas.

```python
students = {
    "name": ['giri', 'raju', 'ramu', 'rani', 'dhanu','hari', 'sita', 'gita', 'laxmi', 'krishna','john','hama'],
    "age": [25, 22, 52, 24, 30, 28, 26, 31, 29, 33, 27, 32],
    "marks": [85, 90, 78, 92, 88, 95, 87, 91, 89, 94, 86, 93]
}
```

```python
df = pd.DataFrame(students) # convert dictionary to dataframe
print(df)
```

```python
df = pd.DataFrame(students, index=["s1", "s2", "s3", "s4", "s5", "s6", "s7", "s8", "s9", "s10", "s11", "s12"]) # convert dictionary to dataframe with custom index
print(df)
```

```python
print(df.loc['s2']) # loc is used to access a group of rows and columns by label(s) or a boolean array.
```

```python
print(df.iloc[2]) # iloc is used to access a group of rows and columns by integer position(s)
```

```python
df["medium"]=["eng","tam","phy","math","chy","eng","tam","phy","math","chy","eng","tam"] # add new column to dataframe
print(df)   
```

```python
s1 = pd.DataFrame([
    {"name": "kumar", "age": 25, "marks": 85, "medium": "bio"},
    {"name": "MJ", "age": 20, "marks": 90, "medium": "eng"}
    ], index=["s13", "s14"]) # create dataframe with single row and custom index
```

```python
df =pd.concat([df, s1] )
print(df)
```
