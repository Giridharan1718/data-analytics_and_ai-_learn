
# Aggregate Functions in Pandas | Ep-7 Pandas for detail

Aggregate functions are used to summarize data by computing a single value from multiple rows. Common aggregate functions in Pandas include:

- **sum()** - Calculate the sum of values
- **mean()** - Calculate the average
- **median()** - Calculate the middle value
- **std()** - Calculate standard deviation
- **min()** - Find minimum value
- **max()** - Find maximum value
- **count()** - Count non-null values
- **agg()** - Apply multiple functions at once

## Basic Usage

Use these functions directly on a DataFrame or Series to get summary statistics for your data.

## Groupby with Aggregation

Combine `groupby()` with aggregate functions to summarize data by categories and gain deeper insights into your dataset.

```python
import pandas as pd
dos=(r"D:\\leanring\\aids\\learning\\pandes\\datasets\\iris.json")
df = pd.read_json(dos)
df 
```

```python
print(df.mean(numeric_only=True)) 

```

```python
print(df.max(numeric_only=True)) 

```

```python
print(df.min(numeric_only=True))
```

```python
print(df.median(numeric_only=True))
```

```python
print(df['petal.length'].mean()) 
print(df['petal.length'].max())
print(df['petal.length'].min())
```

```python

print(df['petal.length'].sum())
```

```python

group=df.groupby('variety')
print(group)
```

```python
print(group.mean(numeric_only=True))
```

# Data Cleaning in Pandas | Ep-8 Pandas

Data cleaning helps make your dataset accurate, consistent, and ready for analysis.

## Common Data Cleaning Steps

- **Handle missing values** using `fillna()`, `dropna()`
- **Drop unnecessary columns** using `drop()`
- **Rename or standardize values** using `replace()`
- **Convert text format** (e.g., lowercase) using string methods
- **Fix data types** using `astype()`
- **Remove duplicate rows** using `drop_duplicates()`

## Workflow Used in This Notebook

- Loaded a second dataset (`iris2.json`)
- Dropped `sepal.length` to create `df1`
- Filled missing values in `sepal.width`
- Replaced category values in `variety` (`Setosa` → `Sta`, `Virginica` → `Vir`)
- Converted `variety` text to lowercase
- Converted `sepal.length` to integer type
- Removed duplicate records

## Why Cleaning Matters

Clean data improves:
- reliability of summary statistics
- grouping and aggregation results
- model performance and interpretation

```python

ios=(r"D:\\leanring\\aids\\learning\\pandes\\datasets\\iris2.json")
df = pd.read_json(ios)
df 
```
