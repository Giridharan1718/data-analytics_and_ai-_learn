## Selection in Pandas | Ep-5 Pandas

### Overview
Selection in Pandas is used to access specific rows, columns, or subsets of a DataFrame or Series. It helps you inspect and work with only the data you need.

### Key Concepts Covered

**Column Selection**
- Select a single column with `df["column"]`
- Select multiple columns with `df[["col1", "col2"]]`

**Row Selection**
- Use `.loc[]` to select rows by label
- Use `.iloc[]` to select rows by integer position

**Series Selection**
- A single column returns a Series
- Example: `city = df["City"]`

**Common Selection Methods**
- `df.head()` → first few rows
- `df.loc[row_label]` → row by label
- `df.iloc[row_index]` → row by position
- `df.iloc[row_start:row_end]` → slice rows

### Practical Examples
- `df["City"]` → selects the City column
- `df[['Customer Id', 'City', 'First Name', 'Last Name']]` → selects specific columns
- `df.loc[5]` → selects row with label 5
- `df.iloc[5]` → selects the 6th row
- `df.iloc[0:5, 0:3]` → selects first 5 rows and first 3 columns

### Notes
- Use `.loc[]` when working with index labels
- Use `.iloc[]` when working with row and column positions
- Selection is a core skill for filtering, cleaning, and analyzing data

## Filtering in Pandas | Episode 6

### Overview
Filtering is a fundamental technique in data analysis that allows you to subset your data based on specific conditions. In this episode, we explored various methods to filter Pandas DataFrames using boolean indexing and comparison operators.

### Key Concepts Covered

**Boolean Indexing**
- Filter rows where a condition is True
- Returns only the rows that match your criteria
- Syntax: `df[df["column"] condition]`

**Comparison Operators**
- `>` Greater than
- `<` Less than
- `==` Equal to
- `!=` Not equal to
- `<=` Less than or equal to
- `>=` Greater than or equal to

**Combining Multiple Conditions**
- Use `&` (AND) to combine conditions: both must be True
- Use `|` (OR) to combine conditions: at least one must be True
- Use `~` (NOT) to negate a condition
- Always wrap each condition in parentheses

### Practical Examples from the Iris Dataset

1. **Single Condition (Greater Than)**
    - `df[df["sepal.length"] > 5.0]` → 118 rows with sepal length greater than 5.0

2. **Single Condition (Less Than or Equal)**
    - `df[df["sepal.length"] <= 5.0]` → 32 rows with sepal length ≤ 5.0

3. **Equality Filter**
    - `df[df['variety'] == "Setosa"]` → 50 Setosa flowers

4. **Inequality Filter**
    - `df[df['variety'] != "Setosa"]` → 100 non-Setosa flowers (Versicolor and Virginica)

5. **Multiple Conditions (AND)**
    - `df[(df['variety'] == "Setosa") & (df['petal.length'] <= 5.0)]` → 50 Setosa flowers with petal length ≤ 5.0

### Best Practices
- Always use parentheses around each condition when combining with `&` or `|`
- Check your data with `.head()` or `.info()` before filtering
- Use `!=` for "not equal to" comparisons
- Use `~` with parentheses for negating entire conditions: `df[~(df["variety"] == "Setosa")]`
