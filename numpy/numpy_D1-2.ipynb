# Introduction to NumPy

NumPy (Numerical Python) is a fundamental library for scientific computing in Python. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays efficiently.

## Key Features

- **N-dimensional Array Object (`ndarray`)**: Core data structure for storing elements of the same type.
- **Broadcasting**: Perform arithmetic operations on arrays of different shapes.
- **Vectorization**: Apply operations to entire arrays without explicit loops, leading to faster code.
- **Mathematical Functions**: Includes functions for linear algebra, statistics, Fourier analysis, and more.
- **Integration with Other Libraries**: Forms the foundation for libraries like Pandas, SciPy, and scikit-learn.

## Why Use NumPy?

- **Performance**: NumPy operations are implemented in C, making them much faster than native Python lists.
- **Convenience**: Simplifies code for mathematical and statistical operations.
- **Memory Efficiency**: Arrays consume less memory compared to Python lists.

## Basic Concepts

### 1. Array Creation

```python
import numpy as np
arr = np.array([1, 2, 3, 4, 5])
```

### 2. Array Attributes

- `shape`: Dimensions of the array
- `dtype`: Data type of elements
- `ndim`: Number of dimensions

### 3. Array Operations

- **Element-wise arithmetic**: `arr1 + arr2`, `arr1 * arr2`
- **Aggregation**: `np.sum(arr)`, `np.mean(arr)`
- **Reshaping**: `arr.reshape((2, 3))`

### 4. Data Types

NumPy supports various data types, such as `int8`, `int16`, `float32`, `bool_`, and `str_`. You can specify the data type when creating an array.

### 5. Slicing and Indexing

NumPy arrays support advanced slicing and indexing, allowing you to access and modify subsets of data efficiently.

## Example

```python
import numpy as np
a = np.array([[1, 2, 3], [4, 5, 6]])
print(a.shape)      # (2, 3)
print(a[0, 1])      # 2
print(a.sum(axis=0)) # [5 7 9]
```

## Summary

NumPy is essential for data analysis, machine learning, and scientific computing in Python. Its efficient array operations and mathematical functions make it a powerful tool for handling numerical data.

## Day 2: Data Types in NumPy

````python
arr = np.array([1, 2, 3, 4, 5], dtype=np.int8) # This line creates a NumPy array with the specified list of integers and explicitly sets the data type of the elements to 8-bit integers (int8). This means that each element in the array will be stored using 8 bits, which can represent values from -128 to 127. If any value in the list exceeds this range, it will wrap around due to overflow.
print(arr)
print(arr.dtype)
````

````python
print(arr)
print(arr.dtype)
print(f"{arr.nbytes} bytes") # This line calculates and prints the total number of bytes consumed by the elements in the NumPy array 'arr'. The 'nbytes' attribute returns the total number of bytes used to store the array's data. Since 'arr' is an array of 5 elements with a data type of int8 (which uses 1 byte per element), the output will be '5 bytes'.
````

````python
arr = np.array([128, 2, 3, 4, 5], dtype=np.int8) # This line creates a NumPy array with the specified list of integers and explicitly sets the data type of the elements to 8-bit integers (int8). Since the value 128 exceeds the maximum value for int8 (127), it will wrap around due to overflow, resulting in -128 being stored in the array instead of 128.
print(arr)
````


# Reason for OverflowError: Python integer 128 out of bounds for int8

The error occurs because the value `128` cannot be represented in the `int8` data type.  
- `int8` can only store integers from **-128 to 127**.
- Assigning `128` to an `int8` array causes an overflow, resulting in an error or unexpected value.

**Solution:**  
Use a larger integer type, such as `int16`, to store values outside the `int8` range.
```

````python
arr = np.array([128, 2010, 300, 400, 500], dtype=np.str_)
print(arr)
print(arr.dtype) 
print(f"{arr.nbytes} bytes")
````



- **Values:** `['128' '2010' '300' '400' '500']`
- **Data type:** `<U4` (Unicode string of length 4)
- **Memory usage:** Each string element uses 16 bytes (4 characters × 4 bytes per Unicode character), so the total size is `arr.nbytes` bytes.

Using `dtype=np.str_` converts all elements to strings, which is useful for storing non-numeric data or mixed types in a single array.


````python
array= np.array(["code io","lanch pad","Giridharan"]) # This line creates a NumPy array with the specified list of strings. Each string in the list will be an element in the array. The output will show the array containing the strings "code io", "lanch pad", and "Giridharan", and the data type will be 'str_'.
print(array)
print(array.dtype)
print(f"{array.nbytes} bytes")
````

````python
array= np.array(["code io","lanch pad","Giridharan"], dtype="<U5") # This line creates a NumPy array with the specified list of strings and explicitly sets the data type to Unicode strings with a maximum length of 5 characters (dtype="<U5"). This means that each string in the array will be truncated to 5 characters if it exceeds that length. The output will show the array with truncated string values, and the data type will be 'str_' with a maximum length of 5.
print(array)
print(array.dtype)
print(f"{array.nbytes} bytes")
````

````python
array= np.array([1,2,True,"lanch pad","Giridharan"]) # This line creates a NumPy array with a mixed list of elements, including integers, a boolean value, and strings. NumPy will upcast the data type to accommodate all types of elements, resulting in an array of strings (dtype='<U11') where each element is converted to its string representation. The output will show the array with string values, and the data type will be 'str_' with a maximum length of 11 characters.
print(array)
print(array.dtype)
print(f"{array.nbytes} bytes")
````

````python
array= np.array([1,2,True,"lanch pad","Giridharan"],dtype=np.object_) # most not use object 
print(array)
print(array.dtype)
print(f"{array.nbytes} bytes")
````

````python
array= np.array([True,False,True,True], dtype=np.bool_) # This line creates a NumPy array with the specified list of boolean values and explicitly sets the data type to boolean (bool_). Each element in the array will be stored as a boolean value (True or False). The output will show the array with boolean values, and the data type will be 'bool_'.
print(array)
print(array.dtype)
print(f"{array.nbytes} bytes")
````

````python
array =np.array([1, 2, 3, 4, 5])
print(array)
print(array.dtype)
print(f"{array.nbytes} bytes")
````

````python
array=array.astype(np.float16) # This line converts the data type of the elements in the NumPy array 'array' to 16-bit floating-point (float16) using the 'astype' method. This means that each element in the array will be stored using 16 bits, which can represent a smaller range of values with less precision compared to the default 64-bit floating-point (float64). The output will show the array with float16 precision, and the data type will be 'float16'.
print(array)
print(array.dtype)
print(f"{array.nbytes} bytes")
````

````python
array=array.astype(np.bool_) # This line converts the data type of the elements in the NumPy array 'array' to boolean (bool_) using the 'astype' method. This means that each element in the array will be evaluated for its truthiness, where non-zero values will be converted to True and zero values will be converted to False. The output will show the array with boolean values, and the data type will be 'bool_'.
print(array)
print(array.dtype)
print(f"{array.nbytes} bytes")
````
