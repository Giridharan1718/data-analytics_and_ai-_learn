## Linear Algebra  
Intuition

AI models = matrix transformations

- **Data** → matrix  
- **Weights** → matrix  
- **Output** = matrix multiplication

Core operation: Matrix Multiplication

\[
C_{ij} = \sum_k A_{ik} B_{kj}
\]

Where:
- \(A\) and \(B\) are matrices

```python
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

C = A @ B
print(C)
```

```python
np.dot(A, B)          # same as @
```

```python
np.linalg.inv(A)      # inverse
```

```python
np.linalg.det(A)      # determinant
```

```python
np.linalg.eig(A)      # eigenvalues
```
