```python

a = np.arange(15).reshape(3, 5) # reshape the array to 3 rows and 5 columns
print(a)
print(a.shape) # print the shape of the array
print(a.ndim) # print the number of dimensions
print(a.dtype.name) # print the data type name
print(a.itemsize) # print the size of each element
print(a.size) # print the total number of elements
print(type(a)) # print the type of the array
b = np.array([6, 7, 8]) 
print(b) # print the array
print(type(b)) # print the type of the array
```
